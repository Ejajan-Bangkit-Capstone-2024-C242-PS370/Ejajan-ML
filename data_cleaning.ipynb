```python
dataset=data.copy()
columns=['Name', 'RecipeIngredientParts',
         'Calories', 'FatContent', 'SaturatedFatContent',
         'CholesterolContent', 'SodiumContent', 'CarbohydrateContent',
         'FiberContent','SugarContent','ProteinContent'
         ]
dataset=dataset[columns]
```

```python
max_Calories=2650
max_daily_fat=100
max_daily_Saturatedfat=13
max_daily_Cholesterol=300
max_daily_Sodium=2300
max_daily_Carbohydrate=325
max_daily_Fiber=40
max_daily_Sugar=40
max_daily_Protein=200
max_list=[max_Calories,
          max_daily_fat,
          max_daily_Saturatedfat,
          max_daily_Cholesterol,
          max_daily_Sodium,
          max_daily_Carbohydrate,
          max_daily_Fiber,
          max_daily_Sugar,
          max_daily_Protein]
```

```python
extracted_data=dataset.copy()
for column,maximum in zip(extracted_data.columns[6:15],max_list):
    extracted_data=extracted_data[extracted_data[column]<maximum]
```

```python
extracted_data.shape
```

```python
extracted_data = extracted_data.dropna(axis=0)
```

```python
extracted_data.shape
```

```python
extracted_data.head()
```

```python
def convert_to_list(r_string):
    # Menghapus `c(` dan tanda kurung di awal dan akhir
    cleaned = r_string.replace('c(', '').strip(')')
    # Memisahkan elemen berdasarkan koma dan menghapus spasi atau tanda kutip
    return ', '.join([item.strip().strip('"') for item in cleaned.split(',')])

# Terapkan fungsi ke kolom 'ingredients'
extracted_data['Ingredients'] = extracted_data['RecipeIngredientParts'].apply(convert_to_list)
```

```python
extracted_data.head()
```

```python
detect_allergen = extracted_data['Ingredients'].unique()
print(detect_allergen)
```

```python
# Define allergen categories and keywords
allergen_keywords = {
    "Dairy": ["milk", "cheese", "butter", "cream", "yogurt"],
    "Nuts": ["pecan", "almond", "peanut", "hazelnut"],
    "Gluten": ["flour", "bread", "oats", "barley", "wheat"],
    "Eggs": ["egg", "egg whites"],
    "Soy": ["soy", "tofu", "soy sauce"],
    "Seafood": ["fish", "shrimp", "crab", "shellfish"],
    "Sesame": ["sesame", "tahini"],
    "Sulphites": ["sulphites", "vinegar"],
    "Legumes": ["peanuts", "lentils", "beans"]
}

def potential_allergens(ingredients):
    allergens = []
    for key, values in allergen_keywords.items():
        for value in values:
            if value in ingredients:
                allergens.append(key)
    return ", ".join(allergens) if allergens else "None"

# Application to dataset
extracted_data['Allergens'] = extracted_data['Ingredients'].apply(potential_allergens)

# View the dataset with allergens
extracted_data.head()
```

```python
extracted_data.to_csv('cleaned_recipes.csv', index=False)
```
